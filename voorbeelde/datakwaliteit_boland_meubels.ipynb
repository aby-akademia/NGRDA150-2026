# DEEL 1: IDENTIFISEER ONTBREKENDE WAARDES

Ontbrekende waardes is data wat nie vasgelê is nie.

```python
# Tel ontbrekende waardes per kolom
ontbrekende_waardes = data.isnull().sum()

print("ONTBREKENDE WAARDES PER KOLOM:")
print("="*60)
for kolom, aantal in ontbrekende_waardes.items():
    if aantal > 0:
        persentasie = (aantal / len(data)) * 100
        print(f"⚠️ {kolom}: {aantal} ontbrekend ({persentasie:.1f}%)")
    else:
        print(f"✅ {kolom}: Geen ontbrekende waardes")
print("="*60)

# Totaal
totaal_ontbrekend = ontbrekende_waardes.sum()
print(f"\nTOTAAL ONTBREKENDE WAARDES: {totaal_ontbrekend}")
```

### Wys rye met ontbrekende waardes

```python
# Vind rye waar enige waarde ontbreek
rye_met_ontbrekende = data[data.isnull().any(axis=1)]

print(f"Aantal rye met ontbrekende waardes: {len(rye_met_ontbrekende)}")
print("\nEerste 10 rye met ontbrekende waardes:")
rye_met_ontbrekende.head(10)
```

### 📊 Visualiseer ontbrekende waardes

```python
# Skep staafgrafiek van ontbrekende waardes
ontbrekende_data = ontbrekende_waardes[ontbrekende_waardes > 0]

if len(ontbrekende_data) > 0:
    plt.figure(figsize=(10, 5))
    ontbrekende_data.plot(kind='bar', color='#E63946')
    plt.title('Ontbrekende waardes per kolom', fontsize=14, fontweight='bold')
    plt.xlabel('Kolom', fontsize=12)
    plt.ylabel('Aantal ontbrekende waardes', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("✅ Geen ontbrekende waardes om te visualiseer nie.")
```

### 💡 INTERPRETASIE: Impak van ontbrekende waardes

**Vir Boland Meubels:**
- **Ontbrekende produknaam:** Kan nie verkope per produk analiseer nie
- **Ontbrekende kosprys:** Kan nie winsmargin bereken nie
- **Ontbrekende produkkategorie:** Kan nie verkope per kategorie vergelyk nie

# DEEL 2: IDENTIFISEER DUPLIKATE

Duplikaatrekords is transaksies wat meer as een keer verskyn.

```python
# Tel duplikate
aantal_duplikate = data.duplicated().sum()

print("DUPLIKAAT ANALISE:")
print("="*60)
print(f"Aantal duplikaatrekords: {aantal_duplikate}")

if aantal_duplikate > 0:
    persentasie = (aantal_duplikate / len(data)) * 100
    print(f"Persentasie van datastel: {persentasie:.2f}%")
    print("="*60)
else:
    print("✅ Geen duplikate gevind nie.")
    print("="*60)
```

### Wys duplikaatrekords

```python
# Wys duplikaatrekords
if aantal_duplikate > 0:
    duplikate = data[data.duplicated(keep=False)].sort_values('transaksie_id')
    print(f"Totale rye (insluitend oorspronklike en duplikate): {len(duplikate)}")
    print("\nEerste 10 duplikaatrekords:")
    display(duplikate.head(10))
    
    # Wys 'n voorbeeld van 'n duplikaat paar
    print("\n⚠️ VOORBEELD VAN DUPLIKAAT:")
    eerste_duplikaat = duplikate.iloc[0]
    print(f"Transaksie ID: {eerste_duplikaat['transaksie_id']}")
    print(f"Produk: {eerste_duplikaat['produk']}")
    print(f"Winkel: {eerste_duplikaat['winkel']}")
    print(f"Datum: {eerste_duplikaat['datum']}")
    print("\nHierdie rekord verskyn meer as een keer in die datastel.")
```

### 💡 INTERPRETASIE: Impak van duplikate

**Vir Boland Meubels:**
- Totale verkope word **oorstaat**
- Bruto wins word **oorskat**
- Voorraadtelling is **onakkuraat**
- Besluite gebaseer op hierdie data sal **foutief** wees

# DEEL 3: IDENTIFISEER ONGELDIGE WAARDES

Ongeldige waardes is data in die verkeerde formaat of wat nie sin maak nie.

## 3.1 Kontroleer datatipes

```python
# Kontroleer datatipes
print("DATATIPES PER KOLOM:")
print("="*60)
for kolom, dtype in data.dtypes.items():
    print(f"{kolom}: {dtype}")
print("="*60)

print("\n⚠️ LET OP:")
print("- Numeriese kolomme moet 'int64' of 'float64' wees")
print("- Teks kolomme moet 'object' wees")
print("- Datums moet 'datetime64[ns]' wees")
```

## 3.2 Identifiseer negatiewe waardes (Logiese Foute)

```python
# Vind negatiewe hoeveelhede
negatiewe_hoeveelheid = data[data['hoeveelheid'] < 0]

print("NEGATIEWE HOEVEELHEDE:")
print("="*60)
if len(negatiewe_hoeveelheid) > 0:
    print(f"⚠️ {len(negatiewe_hoeveelheid)} transaksies met negatiewe hoeveelheid gevind!")
    print("\nRye met negatiewe hoeveelheid:")
    display(negatiewe_hoeveelheid[['transaksie_id', 'datum', 'produk', 'winkel', 'hoeveelheid']])
else:
    print("✅ Geen negatiewe hoeveelhede gevind nie.")
print("="*60)
```

```python
# Vind negatiewe of nul pryse
ongeldige_pryse = data[(data['verkoopsprys'] <= 0) | (data['kosprys'] <= 0)]

print("ONGELDIGE PRYSE (≤ R0):")
print("="*60)
if len(ongeldige_pryse) > 0:
    print(f"⚠️ {len(ongeldige_pryse)} transaksies met ongeldige pryse gevind!")
    print("\nRye met ongeldige pryse:")
    display(ongeldige_pryse[['transaksie_id', 'datum', 'produk', 'verkoopsprys', 'kosprys']])
else:
    print("✅ Geen ongeldige pryse gevind nie.")
print("="*60)
```

## 3.3 Identifiseer onrealistiese winsmarges

```python
# Vind negatiewe winsmarges (kosprys > verkoopsprys)
negatiewe_winsmargin = data[data['winsmarge'] < 0]

print("NEGATIEWE WINSMARGES:")
print("="*60)
if len(negatiewe_winsmargin) > 0:
    print(f"⚠️ {len(negatiewe_winsmargin)} transaksies met negatiewe winsmarge gevind!")
    print("Dit beteken die verkoopsprys is LAER as die kosprys - 'n verlies!")
    print("\nRye met negatiewe winsmarge:")
    display(negatiewe_winsmargin[['transaksie_id', 'datum', 'produk', 'verkoopsprys', 'kosprys', 'winsmarge']].head())
else:
    print("✅ Geen negatiewe winsmarges gevind nie.")
print("="*60)
```

```python
# Vind onrealistiese hoë winsmarges (> 80%)
hoe_winsmargin = data[data['winsmarge'] > 80]

print("ONREALISTIESE HOË WINSMARGES (> 80%):")
print("="*60)
if len(hoe_winsmargin) > 0:
    print(f"⚠️ {len(hoe_winsmargin)} transaksies met winsmargin > 80% gevind!")
    print("Dit is ongewoon hoog en moet ondersoek word.")
    print("\nRye met hoë winsmarge:")
    display(hoe_winsmargin[['transaksie_id', 'datum', 'produk', 'verkoopsprys', 'kosprys', 'winsmarge']].head())
else:
    print("✅ Geen onrealistiese hoë winsmarges gevind nie.")
print("="*60)
```

### 💡 INTERPRETASIE: Impak van ongeldige waardes

**Vir Boland Meubels:**
- **Negatiewe hoeveelhede:** Onmoontlik - duidelike data-invoerfout
- **Pryse van R0:** Verhoed akkurate winsmarginberekening
- **Negatiewe winsmarges:** Dui op foute in kosprys of verkoopsprys
- **Onrealistiese hoë winsmarges:** Moet verifieer - moontlike tikfoute

# DEEL 4: IDENTIFISEER UITSKIETERS

Uitskieters is waardes wat dramaties afwyk van die res van die data.

## 4.1 Beskrywende statistiek

```python
# Beskrywende statistiek vir numeriese kolomme
print("BESKRYWENDE STATISTIEK:")
print("="*60)
data[['verkoopsprys', 'kosprys', 'hoeveelheid', 'winsmarge', 'bruto_wins']].describe()
```

## 4.2 Houer-en-punt stipping vir visuele identifikasie van uitskieters

```python
# Skep bokspotte vir numeriese veranderlikes
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Verkoopsprys
axes[0, 0].boxplot(data['verkoopsprys'].dropna(), vert=False)
axes[0, 0].set_title('Houer-en-punt stipping: Verkoopsprys', fontweight='bold')
axes[0, 0].set_xlabel('Verkoopsprys (R)')

# Kosprys
axes[0, 1].boxplot(data['kosprys'].dropna(), vert=False)
axes[0, 1].set_title('Houer-en-punt stipping: Kosprys', fontweight='bold')
axes[0, 1].set_xlabel('Kosprys (R)')

# Hoeveelheid
axes[1, 0].boxplot(data['hoeveelheid'].dropna(), vert=False)
axes[1, 0].set_title('Houer-en-punt stipping: Hoeveelheid', fontweight='bold')
axes[1, 0].set_xlabel('Hoeveelheid')

# Winsmargin
axes[1, 1].boxplot(data['winsmarge'].dropna(), vert=False)
axes[1, 1].set_title('Houer-en-punt stipping: Winsmarge', fontweight='bold')
axes[1, 1].set_xlabel('Winsmarge (%)')

plt.tight_layout()
plt.show()

print("\n📊 INTERPRETASIE VAN HOUER-EN-PUNT STIPPING:")
print("- Punte buite die 'snorre' is potensiële uitskieters")
print("- Die boks verteenwoordig die middelste 50% van die data")
print("- Die lyn binne die boks is die mediaan")
```

## 4.3 Identifiseer spesifieke uitskieters

```python
# Gebruik IQR metode om uitskieters te vind
def vind_uitskieters(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1
    
    ondergrens = Q1 - 1.5 * IQR
    bogrens = Q3 + 1.5 * IQR
    
    uitskieters = df[(df[kolom] < ondergrens) | (df[kolom] > bogrens)]
    return uitskieters

# Vind uitskieters vir verkoopsprys
prys_uitskieters = vind_uitskieters(data, 'verkoopsprys')

print("UITSKIETERS: VERKOOPSPRYS")
print("="*60)
if len(prys_uitskieters) > 0:
    print(f"⚠️ {len(prys_uitskieters)} uitskieters in verkoopsprys gevind!")
    print("\nTop 5 hoogste pryse (moontlike uitskieters):")
    display(prys_uitskieters.nlargest(5, 'verkoopsprys')[['transaksie_id', 'datum', 'produk', 'verkoopsprys', 'winkel']])
else:
    print("✅ Geen uitskieters in verkoopsprys gevind nie.")
print("="*60)
```

```python
# Vind uitskieters vir hoeveelheid
hoeveelheid_uitskieters = vind_uitskieters(data, 'hoeveelheid')

print("UITSKIETERS: HOEVEELHEID")
print("="*60)
if len(hoeveelheid_uitskieters) > 0:
    print(f"⚠️ {len(hoeveelheid_uitskieters)} uitskieters in hoeveelheid gevind!")
    print("\nTransaksies met ongewone hoeveelhede:")
    display(hoeveelheid_uitskieters.nlargest(5, 'hoeveelheid')[['transaksie_id', 'datum', 'produk', 'hoeveelheid', 'winkel']])
    
    print("\n💡 MOONTLIKE VERKLARING:")
    print("- Groot bestellings (bv. hotel of besigheid)")
    print("- Data-invoerfout (bv. 50 i.p.v. 5)")
    print("→ Elke uitskieter moet INDIVIDUEEL ondersoek word")
else:
    print("✅ Geen uitskieters in hoeveelheid gevind nie.")
print("="*60)
```

### 💡 INTERPRETASIE: Impak van uitskieters

**Vir Boland Meubels:**
- **Uitskieters kan EG wees:** Groot korporatiewe bestelling (hotel bestel 50 beddens)
- **Uitskieters kan FOUTE wees:** Tikfout (verkoopsprys R100 000 i.p.v. R10 000)
- **Impak op statistiek:** Uitskieters verteken gemiddeldes en totale
- **Aksie:** Elke uitskieter moet ondersoek word - moenie outomaties verwyder nie!

# DEEL 5: DOKUMENTASIE VAN DATAKWALITEITSKWESSIES

Nou dokumenteer ons al die kwessies in 'n professionele formaat.

```python
# Skep dokumentasietabel
dokumentasie = []

# 1. Ontbrekende waardes
for kolom, aantal in ontbrekende_waardes.items():
    if aantal > 0:
        dokumentasie.append({
            'Kwessie': f'Ontbrekende {kolom}',
            'Besonderhede': f'{aantal} rye met ontbrekende {kolom}',
            'Impak': f'Kan nie volledige analise op {kolom} doen nie',
            'Moontlike Oplossing': f'Vul met gemiddelde/modus of kontak dataverskaffer'
        })

# 2. Duplikate
if aantal_duplikate > 0:
    dokumentasie.append({
        'Kwessie': 'Duplikaatrekords',
        'Besonderhede': f'{aantal_duplikate} duplikaatrekords gevind',
        'Impak': 'Totale verkope en wins word oorstaat',
        'Moontlike Oplossing': 'Verwyder duplikate, behou slegs eerste rekord'
    })

# 3. Negatiewe hoeveelhede
if len(negatiewe_hoeveelheid) > 0:
    for idx, row in negatiewe_hoeveelheid.iterrows():
        dokumentasie.append({
            'Kwessie': 'Negatiewe hoeveelheid',
            'Besonderhede': f'Ry {idx}: Hoeveelheid = {row["hoeveelheid"]}',
            'Impak': 'Onmoontlike waarde - ondermyn data-akkuraatheid',
            'Moontlike Oplossing': 'Stel op 0 of verifieer met dataverskaffer'
        })

# 4. Ongeldige pryse
if len(ongeldige_pryse) > 0:
    for idx, row in ongeldige_pryse.iterrows():
        dokumentasie.append({
            'Kwessie': 'Ongeldige prys',
            'Besonderhede': f'Ry {idx}: Verkoopsprys = R{row["verkoopsprys"]}, Kosprys = R{row["kosprys"]}',
            'Impak': 'Verhoed akkurate winsmarginberekening',
            'Moontlike Oplossing': 'Vervang met gemiddelde prys of ondersoek'
        })

# 5. Negatiewe winsmarges
if len(negatiewe_winsmargin) > 0:
    for idx, row in negatiewe_winsmargin.head(3).iterrows():  # Beperk tot eerste 3
        dokumentasie.append({
            'Kwessie': 'Negatiewe winsmarge',
            'Besonderhede': f'Ry {idx}: Winsmargin = {row["winsmarge"]:.2f}%',
            'Impak': 'Dui op verlies - moontlike prys- of kosprys-fout',
            'Moontlike Oplossing': 'Verifieer kosprys en verkoopsprys met dataverskaffer'
        })

# 6. Uitskieters
if len(prys_uitskieters) > 0:
    hoogste = prys_uitskieters.nlargest(2, 'verkoopsprys')  # Top 2
    for idx, row in hoogste.iterrows():
        dokumentasie.append({
            'Kwessie': 'Uitskieter (verkoopsprys)',
            'Besonderhede': f'Ry {idx}: Verkoopsprys = R{row["verkoopsprys"]:,.2f}',
            'Impak': 'Verteken gemiddelde prys en statistieke',
            'Moontlike Oplossing': 'Ondersoek: Eg (groot bestelling) of fout (tikfout)?'
        })

# Skep DataFrame
dokumentasie_df = pd.DataFrame(dokumentasie)

print("="*80)
print("DATAKWALITEIT DOKUMENTASIE - BOLAND MEUBELS & TOESTELLE")
print("="*80)
print(f"\nTotale aantal kwessies geïdentifiseer: {len(dokumentasie_df)}\n")
display(dokumentasie_df)
print("="*80)
```

### Stoor dokumentasie na CSV

```python
# Stoor dokumentasie
dokumentasie_df.to_csv('boland_meubels_datakwaliteit_dokumentasie.csv', index=False)
print("✅ Dokumentasie gestoor na: boland_meubels_datakwaliteit_dokumentasie.csv")
```
