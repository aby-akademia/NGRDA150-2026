### Data
Ons gebruik nog dieselfde datastel as in Leereenheid 2

```python
# Laai die Boland Meubels data
url = "https://raw.githubusercontent.com/aby-akademia/NGRDA150-2026/main/datastelle/le2_boland_meubels_verkope.csv"
data = pd.read_csv(url)

print("Boland Meubels data is gelaai!")
```

Ons kyk eerste na die eerste vyf rye van die data. Dit gee 'n vinnige oorsig op hoe die data lyk.


```python
# 1. Kyk na die eerste paar rye
print(data.head())
```

Daarna kyk ons na nuttige inligting: hoeveel rye en kolomme is daar, watter datatipes word gebruik, en is daar ontbrekende waardes. Dit is goed om die struktuur van die data te verstaan.


```python
# 2. Verstaan die struktuur
print(data.info())
```

Derde bereken ons die beskrywende statistiek vir al die numeriese kolomme: dinge soos gemiddeld, mediaan, minimum, maksimum, ensovoorts.

```python
# 3. Basiese beskrywende statistiek
print(data.describe())
```
