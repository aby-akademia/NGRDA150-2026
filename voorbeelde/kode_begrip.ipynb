# Biblioteek gebruik
Ons gaan een Python-biblioteke gebruik:

pandas: Om CSV data in te lees

```python
import pandas as pd

# Laai die verkoopsdata vanaf die internet
url = "https://raw.githubusercontent.com/aby-akademia/NGRDA150-2026/main/datastelle/le2_boland_meubels_verkope.csv"
data = pd.read_csv(url)
print("data is gelaai")

# Bereken die wins vir elke item
data["wins"] = data["verkoopsprys"] - data["kosprys"]
print(f"wins = {data["wins"]}")
```

```python
# Tel al die winste bymekaar om die totale wins te kry
totale_wins = data["wins"].sum()

# Vertoon die resultaat
print(f"Resultaat: R{totale_wins:.2f}")
```

Vraag: Wat presies bereken hierdie kode?

  B. Die totale wins in rand oor alle transaksies

  C. Die persentasie wins teenoor verkope
  
  D. Die hoogste wins wat in enige transaksie behaal is
  
  E. Die aantal transaksies wat winsgewend was

**Korrekte antwoord: B**

```python
# Vind die hoogste wins uit al die transaksies
hoogste_wins = data["wins"].max()

# Vertoon die resultaat
print(f"Resultaat: R{hoogste_wins:.2f}")
```
