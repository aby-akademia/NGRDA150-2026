In hierdie notebook word die basiese beginsels van Python bekendgestel aan die hand van werklike syfers uit die Boland Meubels-gevallestudie. Elke afdeling dek 'n nuwe konsep met 'n praktiese voorbeeld.

---
## 1. Kommentaar in kode

In Python kan ons notas in ons kode skryf deur 'n `#`-simbool te gebruik. Enige teks na die `#` word deur Python geïgnoreer, dit is slegs vir die leser van die kode bedoel. Dit is 'n goeie gewoonte om kode te dokumenteer sodat ander (en jouself later) kan verstaan wat die kode doen.

```python
# Dit is 'n kommentaar, Python sal hierdie reël ignoreer
# Die volgende reël druk 'n boodskap op die skerm
print("Welkom by die Boland Meubels Data-Analise Projek")
```

---
## 2. Die print() Funksie

Die `print()` funksie word gebruik om inligting op die skerm te wys. Dit is een van die mees basiese en nuttigste instruksies in Python. Ons kan teks (in aanhalingstekens) of syfers druk.

```python
# Druk 'n teksboodskap
print("Boland Meubels & Toestelle")

# Druk 'n getal
print(2800000)

# Druk teks en 'n getal saam
print("Maandelikse inkomste: R", 2800000)
```

---
## 3. Veranderlikes

In Python stoor ons inligting in *veranderlikes*. 'n Veranderlike is 'n naam wat ons aan 'n waarde gee. Dit stel ons in staat om dieselfde waarde op verskeie plekke in ons kode te gebruik sonder om dit elke keer oor te tik.

Die `=` simbool word gebruik om 'n waarde aan 'n veranderlike toe te ken, dit beteken nie "is gelyk aan" soos in wiskunde nie, maar eerder "ken hierdie waarde toe aan hierdie naam".

```python
# Veranderlikes vir Boland Meubels se finansiële inligting
maandelikse_inkomste = 2800000
aantal_winkels = 3
hoofkantoor = "Stellenbosch"

# Ons kan nou die veranderlikes gebruik
print("Hoofkantoor:", hoofkantoor)
print("Aantal winkels:", aantal_winkels)
print("Maandelikse inkomste: R", maandelikse_inkomste)
```

---
## 4. Datatipes

In Python het verskillende soorte data verskillende *datatipes*. Die drie mees algemene datatipes wat ons in hierdie kursus sal gebruik is:

- **int** (integer) — heelgetalle, bv. `3` of `2800000`
- **float** (breukgetal) — desimale getalle, bv. `0.35` of `28.5`
- **str** (string) — teks, bv. `"Stellenbosch"` of `"Boland Meubels"`

Ons kan die datatipe van 'n veranderlike nagaan met die `type()` funksie.

```python
# Voorbeelde van verskillende datatipes
aantal_winkels = 3                  # int
winstmarge = 0.28                   # float
naam = "Boland Meubels & Toestelle" # str

# Nagaan van datatipes
print(type(aantal_winkels))
print(type(winstmarge))
print(type(naam))
```

---
## 5. Basiese bewerkings

Python kan gebruik word as 'n kragtige sakrekenaar. Die basiese wiskundige bewerkings in Python is:

| Bewerking       | Simbool | Voorbeeld    |
|-----------------|---------|-------------|
| Optelling       | `+`     | `5 + 3`     |
| Aftrekking      | `-`     | `5 - 3`     |
| Vermenigvuldiging | `*`   | `5 * 3`     |
| Deling          | `/`     | `5 / 3`     |

```python
# Boland Meubels se finansiële berekeninge
maandelikse_inkomste = 2800000
vorige_winstmarge = 0.35
huidige_winstmarge = 0.28

# Berekening van wins
vorige_wins = maandelikse_inkomste * vorige_winstmarge
huidige_wins = maandelikse_inkomste * huidige_winstmarge

# Berekening van die verlies in wins
verlies = vorige_wins - huidige_wins

# Resultate
# Die {:,.2f} beteken: gebruik kommas as duisendskeidingsteken, en wys twee desimale
print(f"Vorige maandelikse wins:     R{vorige_wins:,.2f}")
print(f"Huidige maandelikse wins:    R{huidige_wins:,.2f}")
print(f"Verlies in maandelikse wins: R{verlies:,.2f}")
```

```python
# Verdeling van inkomste per winkel
stellenbosch_aandeel = 0.50  # 50% van inkomste
paarl_aandeel = 0.30         # 30% van inkomste
worcester_aandeel = 0.20     # 20% van inkomste

stellenbosch_inkomste = maandelikse_inkomste * stellenbosch_aandeel
paarl_inkomste = maandelikse_inkomste * paarl_aandeel
worcester_inkomste = maandelikse_inkomste * worcester_aandeel

print(f"Stellenbosch inkomste: R{stellenbosch_inkomste:,.2f}")
print(f"Paarl inkomste:        R{paarl_inkomste:,.2f}")
print(f"Worcester inkomste:    R{worcester_inkomste:,.2f}")
```
