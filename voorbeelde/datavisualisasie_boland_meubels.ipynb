## Stap 1: Laai die nodige biblioteke

Ons gebruik drie hoofbiblioteke:
- **pandas** - vir datahantering
- **matplotlib** - vir basiese grafieke
- **seaborn** - vir gevorderde en mooier grafieke

```python
# Laai biblioteke
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Stel visualisasie styl
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
    
print("Biblioteke suksesvol gelaai!")
```

## Stap 2: Laai die Boland Meubels datastel

Ons laai die transaksiedata vir die afgelope 18 maande.

```python
# Laai die Boland Meubels verkopedata direk vanaf GitHub
url = "https://raw.githubusercontent.com/aby-akademia/NGRDA150-2026/main/datastelle/le3_boland_meubels_verkope_skoon.csv"

# Laai die data
data = pd.read_csv(url)

# Skakel datum kolom na 'datetime' formaat
data['datum'] = pd.to_datetime(data['datum'])

# Wys die eerste paar rye
print("Eerste 5 rye van die datastel:")
print(data.head())

print(f"\nTotale aantal transaksies: {len(data)}")
print(f"Datumreeks: {data['datum'].min()} tot {data['datum'].max()}")
```

## Stap 3: Basiese verkenning

Voor ons visualisasies skep, kyk ons na basiese inligting oor die data.

```python
# Wys basiese statistieke
print("Basiese statistieke:")
print(data[['verkoopsprys', 'kosprys', 'hoeveelheid', 'winsmarge']].describe())

print("\nVerdeling per winkel:")
print(data['winkel'].value_counts())

print("\nVerdeling per produkkategorie:")
print(data['produkkategorie'].value_counts())
```

---

# DEEL 1: STAAFGRAFIEKE

Staafgrafieke gebruik ons om kategoriese data te vergelyk.

## Visualisasie 1: Totale verkope per winkel

```python
# Bereken totale verkope per winkel
verkope_per_winkel = data.groupby('winkel')['bruto_wins'].sum().sort_values(ascending=False)

# Skep staafgrafiek
plt.figure(figsize=(10, 6))
verkope_per_winkel.plot(kind='bar', color=['#2E86AB', '#A23B72', '#F18F01'])
plt.title('Totale bruto wins per winkel (18 Maande)', fontsize=14, fontweight='bold')
plt.xlabel('Winkel', fontsize=12)
plt.ylabel('Bruto Wins (R)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

# Voeg waardes bo-op elke staaf
for i, v in enumerate(verkope_per_winkel):
    plt.text(i, v, f'R{v:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n📊 INTERPRETASIE:")
print(f"- {verkope_per_winkel.idxmax()} genereer die meeste bruto wins: R{verkope_per_winkel.max():,.2f}")
print(f"- {verkope_per_winkel.idxmin()} genereer die minste bruto wins: R{verkope_per_winkel.min():,.2f}")
print(f"- Verskil tussen hoogste en laagste: R{verkope_per_winkel.max() - verkope_per_winkel.min():,.2f}")
```

## Visualisasie 2: Verkope per produkkategorie

```python
# Bereken totale verkope per produkkategorie
verkope_per_kategorie = data.groupby('produkkategorie')['bruto_wins'].sum()

# Skep staafgrafiek met seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=verkope_per_kategorie.index, y=verkope_per_kategorie.values, 
            palette=['#06AED5', '#DD1C1A'])
plt.title('Totale bruto wins per produkkategorie', fontsize=14, fontweight='bold')
plt.xlabel('Produkkategorie', fontsize=12)
plt.ylabel('Bruto wins (R)', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Voeg waardes bo-op elke staaf
for i, v in enumerate(verkope_per_kategorie.values):
    plt.text(i, v, f'R{v:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n📊 INTERPRETASIE:")
print(f"- {verkope_per_kategorie.idxmax()} dra die meeste by tot bruto wins")
print(f"- Bydrae: {(verkope_per_kategorie.max() / verkope_per_kategorie.sum() * 100):.1f}% van totale wins")
```

# DEEL 2: LYNGRAFIEKE

Lyngrafieke gebruik ons om tendense oor tyd te wys.

## Visualisasie 3: Gemiddelde winsmarge oor tyd

```python
# Voeg maand-jaar kolom by
data['maand'] = data['datum'].dt.to_period('M')

# Bereken gemiddelde winsmargin per maand
maandelikse_winsmargin = data.groupby('maand')['winsmarge'].mean()

# Skep lyngrafiek
plt.figure(figsize=(14, 6))
plt.plot(maandelikse_winsmargin.index.astype(str), maandelikse_winsmargin.values, 
         marker='o', linewidth=2, markersize=6, color='#C73E1D')
plt.title('Gemiddelde winsmarge per maand (18 Maande)', fontsize=14, fontweight='bold')
plt.xlabel('Maand', fontsize=12)
plt.ylabel('Winsmarge (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)

# Voeg horisontale lyn by vir gemiddeld
plt.axhline(y=maandelikse_winsmargin.mean(), color='green', linestyle='--', 
            label=f'Gemiddeld: {maandelikse_winsmargin.mean():.2f}%')
plt.legend()

plt.tight_layout()
plt.show()

print("\nINTERPRETASIE:")
print(f"- Hoogste winsmarge: {maandelikse_winsmargin.max():.2f}% in {maandelikse_winsmargin.idxmax()}")
print(f"- Laagste winsmarge: {maandelikse_winsmargin.min():.2f}% in {maandelikse_winsmargin.idxmin()}")
print(f"- Gemiddelde winsmarge: {maandelikse_winsmargin.mean():.2f}%")

# Tendens
if maandelikse_winsmargin.iloc[-1] < maandelikse_winsmargin.iloc[0]:
    print("TENDENS: Winsmarge toon 'n dalende patroon oor die 18 maande")
else:
    print("TENDENS: Winsmarge toon 'n stygende patroon oor die 18 maande")
```

## Visualisasie 4: Winsmargin per winkel oor tyd

```python
# Bereken maandelikse winsmargin per winkel
winsmargin_per_winkel = data.groupby(['maand', 'winkel'])['winsmargin'].mean().unstack()

# Skep lyngrafiek met drie lyne
plt.figure(figsize=(14, 6))
for winkel in winsmargin_per_winkel.columns:
    plt.plot(winsmargin_per_winkel.index.astype(str), winsmargin_per_winkel[winkel], 
             marker='o', linewidth=2, markersize=5, label=winkel)

plt.title('Winsmargin per Winkel oor Tyd', fontsize=14, fontweight='bold')
plt.xlabel('Maand', fontsize=12)
plt.ylabel('Winsmargin (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Winkel', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n📊 INTERPRETASIE:")
for winkel in winsmargin_per_winkel.columns:
    gemiddeld = winsmargin_per_winkel[winkel].mean()
    print(f"- {winkel}: Gemiddelde winsmargin = {gemiddeld:.2f}%")
```

# DEEL 3: SPREIDIAGRAMME

Spreidiagramme gebruik ons om verhoudings tussen twee veranderlikes te ondersoek.

## Visualisasie 5: Verkoopsprys vs. Winsmarge

```python
# Skep spreidiagram
plt.figure(figsize=(10, 6))
plt.scatter(data['verkoopsprys'], data['winsmarge'], alpha=0.5, s=50, color='#6A4C93')
plt.title('Verhouding: Verkoopsprys vs. Winsmarge', fontsize=14, fontweight='bold')
plt.xlabel('Verkoopsprys (R)', fontsize=12)
plt.ylabel('Winsmarge (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Bereken korrelasie
korrelasie = data['verkoopsprys'].corr(data['winsmarge'])

print("\n📊 INTERPRETASIE:")
print(f"- Korrelasie tussen verkoopsprys en winsmarge: {korrelasie:.3f}")
if korrelasie > 0.5:
    print("  Sterk positiewe korrelasie: Duurder produkte het hoër winsmarges")
elif korrelasie < -0.5:
    print("  Sterk negatiewe korrelasie: Duurder produkte het laer winsmarges")
else:
    print("  Swak korrelasie: Geen duidelike verband tussen prys en winsmarge")
```

## Visualisasie 6: Verkoopsprys vs. Winsmarge (per produkkategorie)

```python
# Skep spreidiagram met kleur per kategorie
plt.figure(figsize=(10, 6))
for kategorie in data['produkkategorie'].dropna().unique():
    subset = data[data['produkkategorie'] == kategorie]
    plt.scatter(subset['verkoopsprys'], subset['winsmarge'], 
                alpha=0.6, s=50, label=kategorie)

plt.title('Verkoopsprys vs. Winsmarge per produkkategorie', fontsize=14, fontweight='bold')
plt.xlabel('Verkoopsprys (R)', fontsize=12)
plt.ylabel('Winsmarge (%)', fontsize=12)
plt.legend(title='Kategorie')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n📊 INTERPRETASIE:")
for kategorie in data['produkkategorie'].dropna().unique():
    subset = data[data['produkkategorie'] == kategorie]
    korr = subset['verkoopsprys'].corr(subset['winsmarge'])
    print(f"- {kategorie}: Korrelasie = {korr:.3f}")
```

### 💡 TAAK 3: Identifiseer uitskieters

Kyk na die spreidiagram hierbo. Kan jy enige uitskieters identifiseer?

Vind die produk met die **hoogste verkoopsprys**:

```python
# Vind die rekord met hoogste verkoopsprys
hoogste_prys = data.loc[data['verkoopsprys'].idxmax()]

print("Produk met hoogste verkoopsprys:")
print(f"- Produk: {hoogste_prys['produk']}")
print(f"- Verkoopsprys: R{hoogste_prys['verkoopsprys']:,.2f}")
print(f"- Winsmarge: {hoogste_prys['winsmarge']:.2f}%")
print(f"- Winkel: {hoogste_prys['winkel']}")
print(f"- Datum: {hoogste_prys['datum']}")
print("\nIs hierdie 'n uitskieter of 'n egte transaksie? Dit moet ondersoek word!")
```

# DEEL 4: SAMEVATTING EN KOMMUNIKASIE

## Finale bevindinge vir Pieter van der Merwe

```python
print("="*60)
print("SAMEVATTING: BOLAND MEUBELS & TOESTELLE")
print("Datavisualisasie-analise oor 18 maande")
print("="*60)

print("\n1️⃣ WINKEL PRESTASIE:")
verkope_per_winkel = data.groupby('winkel')['bruto_wins'].sum().sort_values(ascending=False)
for winkel, wins in verkope_per_winkel.items():
    persentasie = (wins / verkope_per_winkel.sum()) * 100
    print(f"   - {winkel}: R{wins:,.0f} ({persentasie:.1f}% van totaal)")

print("\n2️⃣ PRODUKKATEGORIE PRESTASIE:")
verkope_per_kategorie = data.groupby('produkkategorie')['bruto_wins'].sum().sort_values(ascending=False)
for kategorie, wins in verkope_per_kategorie.items():
    persentasie = (wins / verkope_per_kategorie.sum()) * 100
    print(f"   - {kategorie}: R{wins:,.0f} ({persentasie:.1f}% van totaal)")

print("\n3️⃣ WINSMARGE TENDENSE:")
maandelikse_winsmargin = data.groupby('maand')['winsmarge'].mean()
print(f"   - Gemiddelde winsmargin: {maandelikse_winsmargin.mean():.2f}%")
print(f"   - Hoogste maand: {maandelikse_winsmargin.max():.2f}% ({maandelikse_winsmargin.idxmax()})")
print(f"   - Laagste maand: {maandelikse_winsmargin.min():.2f}% ({maandelikse_winsmargin.idxmin()})")

verandering = maandelikse_winsmargin.iloc[-1] - maandelikse_winsmargin.iloc[0]
if verandering < 0:
    print(f"   Winsmarge het gedaal met {abs(verandering):.2f}% oor 18 maande")
else:
    print(f"   Winsmarge het gestyg met {verandering:.2f}% oor 18 maande")

print("\n4️⃣ SLEUTELBEVINDINGE:")
print("   - Stellenbosch-winkel genereer die meeste inkomste")
print("   - Meubels dra meer by tot totale wins as Toestelle")
print("   - Winsmarges toon variasie oor tyd - moontlike seisoenaliteit")
print("   - Geen sterk korrelasie tussen verkoopsprys en winsmarge nie")

print("\n5️⃣ AANBEVELINGS:")
print("   1. Ondersoek waarom sekere maande laer winsmarges het")
print("   2. Fokus op meubels - dit genereer beter winsmarges")
print("   3. Analiseer Worcester-winkel se prestasie in meer detail")
print("   4. Identifiseer uitskieters en verifieer transaksies")

print("\n" + "="*60)
```
