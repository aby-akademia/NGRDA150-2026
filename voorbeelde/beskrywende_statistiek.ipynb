### Biblioteek gebruik
Ons gaan een Python-biblioteek gebruik:

Pandas: Om CSV data in te lees

```python
# Laai die Pandas biblioteek
import pandas as pd

print("Pandas biblioteek is gelaai!")
```

### Data
Ons gebruik nog dieselfde datastel as in Leereenheid 2.

```python
# Laai die Boland Meubels data
url = "https://raw.githubusercontent.com/aby-akademia/NGRDA150-2026/main/datastelle/le2_boland_meubels_verkope.csv"
data = pd.read_csv(url)

print("Boland Meubels data is gelaai!")
```

Bereken die winsmarge en stoor dit in die DataFrame as 'n nuwe kolom.


```python
# 1. Bereken die winsmarge
# Ons trek die kosprys van die verkoopsprys af om die wins te kry,
# deel deur die verkoopsprys en vermenigvuldig met 100 vir die persentasie.
data["winsmarge"] = ((data["verkoopsprys"] - data["kosprys"]) / data["verkoopsprys"]) * 100
print("Winsmarge is bereken!\n")

# kyk na die eerste paar rye
print(data.head())
```

Bereken die beskrywende statistiek vir al die numeriese kolomme: dinge soos gemiddeld, mediaan, minimum, maksimum, ensovoorts.


```python
# 2. Beskrywende statistieke vir alle numeriese kolomme
print("Beskrywende statistieke:")
print(data[["verkoopsprys", "kosprys", "winsmarge"]].describe())
```

Bereken die maatstawwe van sentrale neiging handmatig.

```python
# 3. Bereken handmatig: gemiddeld, mediaan, modus
gemiddelde_winsmarge = data["winsmarge"].mean()
mediaan_winsmarge = data["winsmarge"].median()
modus_winsmarge = data["winsmarge"].mode()[0]
print(f"\nGemiddelde winsmarge: {gemiddelde_winsmarge:.2f}%")
print(f"Mediaan winsmarge: {mediaan_winsmarge:.2f}%")
print(f"Modus winsmarge: {modus_winsmarge:.2f}%")

print("\nSimmetriese verspreiding (nul skuinsheid), Gemiddeld = Mediaan = Modus")
```
